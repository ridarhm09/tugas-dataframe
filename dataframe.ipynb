```python
# 1.  Dengan menggunakan pustaka pandas di Python, buatlah sebuah DataFrame dari 
# data jumlah produksi sampah berdasarkan Kabupaten/Kota di Jawa Barat. Pastikan 
# kolom-kolomnya menyertakan nama Kabupaten/Kota, jumlah produksi sampah (dalam 
# ton), dan tahun pencatatan.
import pandas as pd

# baca file
df_sampah = pd.read_excel('disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx', sheet_name='data')
df_sampah

df_hasil = df_sampah[['nama_kabupaten_kota','jumlah_produksi_sampah','satuan','tahun']]
df_hasil
# df_hasil.to_csv('DataFrame_no 1.csv', index=False)
# df_hasil.to_excel('DataFrame_no 1.xlsx', index=False, sheet_name='nomor 1')
```

```python
# 2. Dari DataFrame yang telah dibuat, hitunglah total produksi sampah di seluruh 
# Kabupaten/Kota di Jawa Barat untuk tahun tertentu. Tampilkan hasilnya.

total = 0

for i, row in df_sampah.iterrows():
    Tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    if Tahun == 2020:
        total += jumlah

hasil=[("2020", total)]

df_hasil = pd.DataFrame(hasil, columns=['Tahun', 'jumlah_produksi_sampah'])
df_hasil
# df_hasil.to_csv('DataFrame_no 2.csv', index=False)
# df_hasil.to_excel('DataFrame_no 2.xlsx', index=False, sheet_name='nomor 2')
```

```python
# 3. Jumlahkan Data Pertahun
data_tahun = {}

for i, row in df_sampah.iterrows():
    Tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']
    if Tahun in data_tahun:
        data_tahun[Tahun] += sampah
    else:
        data_tahun[Tahun] = sampah

data = []
for Tahun, jumlah in data_tahun.items():
    data.append([Tahun, jumlah])


df_hasil = pd.DataFrame(data, columns = ["Tahun","Total sampah"])
df_hasil
# df_hasil.to_csv('DataFrame_no 3.csv', index=False)
# df_hasil.to_excel('DataFrame_no 3.xlsx', index=False, sheet_name='nomor 3')
```

```python
# 4. Jumlahkan data per Kota/Kabupaten per tahun

data = {}

for i, row in df_sampah.iterrows():
    Tahun = row['tahun']
    wilayah = row['nama_kabupaten_kota']
    sampah = row['jumlah_produksi_sampah']

    if wilayah not in data:
        data[wilayah] = {}

    if Tahun in data[wilayah]:
        data[wilayah][Tahun] += sampah
    else:
        data[wilayah][Tahun] = sampah
        
hasil = []
for wilayah, total in data.items():
    for Tahun, jml in total.items():
        hasil.append([wilayah, Tahun, jml])

df_hasil = pd.DataFrame(hasil, columns = ['Wilayah', 'Tahun', 'Jumlah Sampah'])
df_hasil 
# df_hasil.to_csv('DataFrame_no 4.csv', index=False)
# df_hasil.to_excel('DataFrame_no 4.xlsx', index=False, sheet_name='nomor 4')
```
